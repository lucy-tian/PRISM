```python
import pandas as pd
import numpy as np
```

```python
# 0.1, 0.03, 0.01, 0.003, 0.001, 3 × 10−4, 1 × 10−4, 3 × 10−5, 1 × 10−5
for i in ['INI30120', 'INI20030780', 'INI1003063']:
    for p in [0.1, 0.03, 0.01, 0.003, 0.001, 0.0003, 0.0001, 0.00003, 0.00001]: 
        label = str(p).replace('.', '_')
        clump = pd.read_csv(f'clumping/{i}.clumped', delim_whitespace=True)
        gwas = pd.read_csv(f'top_SNPs/{i}.topsnps.tsv', sep='\t')
        gwas = gwas[gwas['ID'].isin(clump['SNP'])]
        gwas = gwas[gwas['P'] <= p]
        gwas[['ID', 'A1', 'BETA']].to_csv(f'{i}_{label}.tsv', sep='\t', index=False)
```
