```python
import pandas as pd
import numpy as np
```

```python
for p in ['INI30120', 'INI20030780', 'INI1003063', 'INI50030700']:
    out = pd.read_csv(f'output/{p}/output.agg.txt.gz', delim_whitespace=True)
    sumstat = pd.read_csv(f'sumstat/{p}_Afr_sumstats.txt.gz', delim_whitespace=True)
    out = out.dropna()
    sumstat = sumstat.dropna()
    keep = sumstat.merge(out[['SNP', 'A1', 'A2']], on=['SNP', 'A1', 'A2'])['SNP']
    
    print(len(out))
    print(len(sumstat))
    print(len(keep))
    
    out[out['SNP'].isin(keep)].to_csv(f'linear_input/{p}_finemapp.txt.gz', sep='\t', index=False)
    sumstat[sumstat['SNP'].isin(keep)].to_csv(f'linear_input/{p}_sumstat.txt.gz', sep='\t', index=False)
```
