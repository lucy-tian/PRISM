```python
SNPs = []
for pop in ['Afr', 'WB']:
    for trait in ["INI1003063", "INI20030780", "INI30120", "INI50030700"]:
        if (pop == 'Afr') and (trait == 'INI1003063'):
            continue
        df = pd.read_csv(f'out/{pop}_{trait}_sbrc.txt', sep='\t')
        SNPs.extend(df['SNP'].tolist())

SNPs = list(set(SNPs))
```
